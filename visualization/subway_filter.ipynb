```python
##### API 가져와보기
import requests
import pandas as pd 
import json
from urllib.parse import quote
from bs4 import BeautifulSoup

# 서울 지역 116곳 번호 뽑기
total_list = []

for i in range(1,117):
    area_list = []
    #지역 코드 값 자동 완성
    area = f"POI{i:03d}"
    # API 키
    api_key = '6e51654f776b353335387168745664'
    # API URL 구성
    url = f'http://openapi.seoul.go.kr:8088/{api_key}/xml/citydata/1/2/{area}'
    # API 요청, 응답
    response = requests.get(url)
    xml_data = response.text
    # xml파싱
    soup = BeautifulSoup(xml_data, 'lxml-xml')

    #명소명
    area_nm = soup.find('AREA_NM').text
    total_list.append(area_nm)
    print(area_nm)
```

```python
import pandas as pd
df = pd.DataFrame(total_list, columns=['지역', '주소','구'])
```

```python
import re
filtered = df[df['지역'].str.contains('역')]['지역']
# 괄호 내용을 제거하고 한글+역 패턴 추출
sub_filter = []
for station in filtered:
    # 괄호 제거
    clean_station = re.sub(r'\([^)]*\)', '', station)
    # 한글 문자열 + '역' 추출
    match = re.search('([가-힣]+역)', clean_station)
    if match:
        sub_filter.append(match.group(0))
    else:
        sub_filter.append(clean_station)
print(sub_filter)
print(len(sub_filter))
```

```python
#리스트에서 역글자 제거
filtered_clean = [name.replace('역', '') for name in sub_filter]
#데이터프레임에 맞게 역이름 변경
replacements = {
    '구로': '구로디지털단지',
    '서울': '서울역',
    '삼' : '역삼',
    '마곡나루':'마곡',
    '용산':'신용산',
    '미아사거리':'미아삼거리'  
}

filtered_clean = [replacements.get(item, item) for item in filtered_clean]
#DataFrame에서도 공백제거
subway_level1231['출발역'] = subway_level1231['출발역'].str.strip()
#필터링
filtered_df = subway_level1231[subway_level1231['출발역'].isin(filtered_clean)]
```

```python
# 1시간 단위 시간대 설정 (30분 단위 컬럼 2개씩 묶기)
time_columns = {
    '10시': ['10시00분', '10시30분'],
    '11시': ['11시00분', '11시30분'],
    '12시': ['12시00분', '12시30분'],
    '13시': ['13시00분', '13시30분'],
    '14시': ['14시00분', '14시30분'],
    '15시': ['15시00분', '15시30분'],
    '16시': ['16시00분', '16시30분'],
    '17시': ['17시00분', '17시30분'],
    '18시': ['18시00분', '18시30분'],

}

# 결과 저장용 빈 데이터프레임 생성
result_columns = ['출발역']
for day in ['주말']:
    for hour in time_columns.keys():
        result_columns.append(f'{day}_{hour}_혼잡도')

result_df = pd.DataFrame(columns=result_columns, data=weekend_df)
result_df
```

```python
# 주말 데이터 처리
for hour, cols in time_columns.items():
    if all(col in weekend_df.columns for col in cols):
        hour_avg = weekend_df.groupby('출발역')[cols].mean().mean(axis=1).round().astype(float)
        
        for station, avg in hour_avg.items():
            idx = result_df[result_df['출발역'] == station].index
            if len(idx) > 0:
                result_df.loc[idx, f'주말_{hour}_혼잡도'] = avg

weekend_df
```

```python
# 1. 공백 제거 및 복사
combined_df['요일구분'] = combined_df['요일구분'].str.strip()
df = combined_df.copy()
df['출발역'] = df['출발역'].str.strip()

# 2. 주말 데이터만 필터링
weekend_df = df[df['요일구분'].isin(['토요일', '일요일'])]

# 3. 1시간 단위 시간대 묶기
time_columns = {
    '10시': ['10시00분', '10시30분'],
    '11시': ['11시00분', '11시30분'],
    '12시': ['12시00분', '12시30분'],
    '13시': ['13시00분', '13시30분'],
    '14시': ['14시00분', '14시30분'],
    '15시': ['15시00분', '15시30분'],
    '16시': ['16시00분', '16시30분'],
    '17시': ['17시00분', '17시30분'],
    '18시': ['18시00분', '18시30분'],
}

# 4. 출발역별로 평균 혼잡도 계산 후 새로운 DF 구성
final_data = {}

for hour, cols in time_columns.items():
    # 해당 시간대 컬럼이 모두 존재할 경우만 처리
    if all(col in weekend_df.columns for col in cols):
        avg = weekend_df.groupby('출발역')[cols].mean().mean(axis=1)
        final_data[f'주말_{hour}_혼잡도'] = avg

# 5. 데이터프레임으로 변환
result_df = pd.DataFrame(final_data)
result_df = result_df.reset_index()  # 출발역을 컬럼으로 복원
```
