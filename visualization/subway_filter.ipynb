```python
import re
# 리스트에서 필터링
filtered = [item for item in total_list if '역' in item]  # 1번 인덱스가 지역명

# 괄호와 패턴 처리
sub_filter = []
for item in filtered:
   station = item[1]  # 지역명 추출
   # 괄호 제거
   clean_station = re.sub(r'\([^)]*\)', '', station)
   # 한글 문자열 + '역' 추출
   match = re.search('([가-힣]+역)', clean_station)
   if match:
       # 원본 리스트 복사 후 수정된 지역명 삽입
       new_item = item.copy()
       new_item[1] = match.group(0)
       sub_filter.append(new_item)
   else:
       sub_filter.append(item)

print(len(sub_filter))
```
